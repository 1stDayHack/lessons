```python
#Define and instantiate function
def mySum(var1,var2,do=True):
    
    #Instantiate result to return if do == False
    result = None
    
    if do:
        result = var1 + var2
        print(result)
    else:
        print("Done nothing.")
        
    return result
```

```python
mySum(5,2,False) #Can't return value because it's not instantiated in the function
```

## Classes

```python
class wrongCalculator():
    """
    Docstring if you like. This is a simple calculator.
    """
    
    def __init__(self,fudge1,fudge2): #Note the self value. Necessary for all function to refer to itself.
        """
        The __init__ is a sfunction that will be called when the class is first initiatied.
        Python will always look for this particular function during instantiation.
        Note that these are also called 'magic functions' or 'dunder functions' as denoted by the
        '__' prefix and suffix.
        """
        
        #Initialize class attributes from instantiation input
        self.f1 = fudge1
        self.f2 = fudge2
        self.check_me_later = "Still here."
        
        print("Initalized")
        
        
    def add(self,x1,x2): #Note the self value again.
        """
        Simple 2 value adder.
        """
        return x1 + x2 + self.f1 #Fudge the value
    
    
    def minus(self,x1,x2): #And again.
        """
        Simple 2 value subtractor.
        """
        return x1 - x2 - self.f2
```

```python
Kalkulator = wrongCalculator() #Opps, did not supply the required variable in __init__
```

```python
Kalkulator = wrongCalculator(2,5) #Much better
```

```python
Kalkulator.add(1,1) #Lols
```

```python
Kalkulator.minus(2,2)
```

## Watch out, it's Inheritance!

In its defense, at small scale and well managed larger scale projects (very well managed), inheritance can actually be very handy and time-space saving.

```python
class rightCalculator(wrongCalculator):
    
    def __init__(self):
        """
        Init func again.
        """
        #Super() will let this child function (rightCalculator) inherit all functions of its parent (wrongCalculator) 
        #We're initializing with fudges == 0 so the calculator will be correct.
        super().__init__(fudge1=0,fudge2=0)
        
    
    #Let's try to create a new function with rightCalculator.
    def multiply(self,x,y):
        """
        Simple multiplier.
        """
        return x * y
```

```python
Kalkulator = rightCalculator()
```

```python
Kalkulator.add(1,1)
```

```python
Kalkulator.minus(2,1)
```

```python
Kalkulator.multiply(2,3) #New function added on top of wrongCalculator
```

```python
Kalkulator.check_me_later #Attributes are inherited too.
```

## Creating a Simple Toy Application

This is a simple application that basically opens an image from a given image path, and return as resized version of it for plotting. However, before being resized, it'll check for whether we should resize it and if we do, what size we should resize it to. <br>

```python
class imgViewer():
    """
    Simple image viewer.
    """
    
    def __init__(self,path,resize=True,rsz_dim=(512,512)):
        """
        Initialization.
        """
        self.path = path
        self.image = PIL.Image.open(path)
        
        #Check for rsz
        if resize:
            self.ori = self.image.copy()
            self.image = self.image.resize(rsz_dim)
        
        
    def size(self):
        """
        Simple function that returns dim of image
        """
        return self.image.size
    
    
    def to_np(self):
        """
        Convert image to np.array for manipulation.
        """
        return np.array(self.image)
            
    
    def display(self,ori=False):
        """
        Simple func to plot.
        """
        
        #Plot original or resized
        if ori:
            plt.imshow(self.ori)
            
        else:
            plt.imshow(self.image)
```
